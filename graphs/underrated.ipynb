```python
import pandas as pd
ratings_df = pd.read_csv(r"D:\Braindead-Team-Egghead\data\ratings.csv")
movie_popularity = (
    ratings_df
    .groupby("movieId")
    .size()
    .reset_index(name="num_ratings")
)
```

```python
movie_ratings = (
    ratings_df
    .groupby("movieId")["rating"]
    .mean()
    .reset_index(name="avg_rating")
)
```

```python
movie_stats = movie_ratings.merge(
    movie_popularity,
    on="movieId"
)
```

```python
import matplotlib.pyplot as plt

plt.scatter(
    movie_stats["num_ratings"],
    movie_stats["avg_rating"],
    alpha=0.5
)
plt.xscale("log")
plt.xlabel("Number of Ratings (log scale)")
plt.ylabel("Average Rating")
plt.title("Popularity Bias Visualization")
plt.show()
```
